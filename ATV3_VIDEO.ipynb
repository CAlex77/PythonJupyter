```python
#CODIGO USADO EM SALA PARA DETECÇÃO DE CORES VERMELHAS NA IMAGEM DO TUCANO. 
#LEVEMENTE MODIFICADO PARA EXIBIR AS IMAGENS NA ABA DO JUPYTER NOTEBOOK, AO INVES DE ABRIR UM TERMINAL.
#CODIGO USADO COMO BASE PARA FAZER AS DETECÇÕES DE CORES NOS VIDEOS

import cv2
import numpy as np
import matplotlib.pyplot as plt

path = 'C:/Users/carlo/TETI/Imagens/'
image = cv2.imread(path + 'tucano3.png')
hsv = cv2.cvtColor(image , cv2.COLOR_BGR2HSV)
limiar_claro = np.array([0, 100, 100])
limiar_escuro = np.array([10, 255, 255])
mask = cv2.inRange(hsv, limiar_claro, limiar_escuro)
output = cv2.bitwise_and(image, image, mask = mask)

# Inverte a ordem dos canais de BGR para RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
output = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)

# Cria uma figura com duas subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 5))

# Exibe a imagem padrão do tucano na primeira subplot
ax1.imshow(image)
ax1.set_title('Imagem Original')

# Exibe a imagem modificada do tucano na segunda subplot
ax2.imshow(output)
ax2.set_title('Cor Detectada')

plt.show()
```

```python
#ESSE CODIGO EXIBE AS CORES VERMELHAS MAS NÃO DETECTA TODAS AS POSSIBILIDADES DE TONS

import cv2
import numpy as np
from IPython.display import display, Image

path = 'C:/Users/carlo/TETI/Videos/'
video_path = path + 'caleido.mp4'

cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print('Erro na leitura do vídeo...')
    exit()

# Valores de limiar claro e limiar escuro para detecção de cores vermelhas
limiar_claro = np.array([0, 100, 100])
limiar_escuro = np.array([10, 255, 255])

while True:
    ret, frame = cap.read()
    
    if not ret:
        print('Finalizou o vídeo.')
        break
        
    # Converter o frame de BGR para HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    # Gerar a máscara para a detecção de cores vermelhas
    mask = cv2.inRange(hsv, limiar_claro, limiar_escuro)
    
    # Aplicar a máscara na imagem original
    output = cv2.bitwise_and(frame, frame, mask=mask)
    
    # Exibir o resultado da detecção de cores vermelhas
    _, img_encoded = cv2.imencode('.jpg', output)
    display(Image(data=img_encoded))
    
cap.release()
```

```python
#ESSE CODIGO DEVE DETECTAR TODAS AS POSSIBILIDADES DE CORES, INCLUSIVE TONS PROXIMOS DE ROSA

import cv2
import numpy as np
from IPython.display import display, Image

path = 'C:/Users/carlo/TETI/Videos/'
video_path = path + 'caleido.mp4'

cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print('Erro na leitura do vídeo...')
    exit()

# Valores de limiar claro e limiar escuro para detecção de cores vermelhas
limiar_claro = np.array([0, 50, 50])
limiar_escuro = np.array([10, 255, 255])

while True:
    ret, frame = cap.read()
    
    if not ret:
        print('Finalizou o vídeo.')
        break
        
    # Converter o frame de BGR para HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    # Gerar a máscara para a detecção de cores vermelhas
    mask1 = cv2.inRange(hsv, np.array([0, 50, 50]), np.array([10, 255, 255]))
    mask2 = cv2.inRange(hsv, np.array([170, 50, 50]), np.array([180, 255, 255]))
    mask = cv2.bitwise_or(mask1, mask2)
    
    # Aplicar a máscara na imagem original
    output = cv2.bitwise_and(frame, frame, mask=mask)
    
    # Exibir o resultado da detecção de cores vermelhas
    _, img_encoded = cv2.imencode('.jpg', output)
    display(Image(data=img_encoded))
    
cap.release()
```
